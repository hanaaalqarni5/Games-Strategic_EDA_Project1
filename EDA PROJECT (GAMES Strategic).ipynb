## DATA CLEANING

```python
#let us remove un-necessary columns
df2.drop(["index","URL","ID","Subtitle","Icon URL","In-app Purchases","Description","Languages","Size","Genres","Original Release Date","Current Version Release Date"],axis=1,inplace=True)
```

```python
df2.head()
```

```python
#re-defining columns names
df2.columns=["name","average_rating","user_rating_count","price","developer","age_rating","genre"]
```

```python
df2.head()
```

```python
# remove null values and update the dataframe
df2 = df2.dropna()
df2.isnull().sum()
```

```python
#remove spaces from columns
df2.columns=[column.strip() for column in df2.columns]
df2.columns
```

```python
#remove duplicate values
df2 = df2.drop_duplicates()
print("Total number of duplicate rows: ", df2.duplicated().sum())
```

```python
#check data type
df.dtypes
```

```python
#change AGE columns attributes to rather child or teens 
df2.age_rating=df2.age_rating.apply(lambda x: str(x).replace("+","") if '' in str(x) else str(x))
df2["age_rating"]=df2["age_rating"].apply(lambda x: str(x).replace('4','Child') if '4' in str(x) else str(x))
df2["age_rating"]=df2["age_rating"].apply(lambda x: str(x).replace('9','Child') if '9' in str(x) else str(x))
df2["age_rating"]=df2["age_rating"].apply(lambda x: str(x).replace('12','Teenager') if '12' in str(x) else str(x))
df2["age_rating"]=df2["age_rating"].apply(lambda x: str(x).replace('17','Teenager') if '17' in str(x) else str(x))
```

```python
#let us add rank column:
df2 = df2.sort_values(by=["user_rating_count","genre"], ascending=False)
df2['rank']=tuple(zip(df2.user_rating_count,df2.genre))
df2['rank']=df2.groupby('user_rating_count',sort=False)['rank'].apply(lambda x : pd.Series(pd.factorize(x)[0])).values
df2.head()
```

```python
#re-set the index from zero
df2.reset_index(inplace=True,drop=True)
df2.head()
```
